### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
NewsList = []
Categories = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    category = d.find_all('em')[0].text
    if category in Categories:
        value = Categories[category] + 1
        Categories.update([(category, value)])
    else:
        Categories.update([(category, 1)] )
    date = d.find(class_="date").text
    subject = d.find_all('a')[-1].text
    NewsList.append([category, date, subject.replace(u'\u3000',u' ')])
SortedList = sorted(NewsList, key = lambda x: x[0]) 
for item in SortedList:
    print(item[0], item[1], item[2])
```

### ③ 哪一個類別的文章最多
